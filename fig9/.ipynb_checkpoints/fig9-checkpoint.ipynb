```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

def generate_default_plot_env():
    sns.set(font_scale=1.4, style="whitegrid")
    matplotlib.rcParams['pdf.fonttype'] = 42
    matplotlib.rcParams['ps.fonttype'] = 42
    matplotlib.rc('font', family='sans-serif') 
    matplotlib.rc('font', serif='Helvetica Neue') 
    matplotlib.rc('text', usetex='false') 
    matplotlib.rcParams.update({'font.size': 22})
    
    plt.rcParams['pdf.fonttype'] = 42
    plt.rcParams['ps.fonttype'] = 42
    plt.rc('font', family='sans-serif') 
    plt.rc('font', serif='Helvetica Neue') 
    plt.rc('text', usetex='false') 
    plt.rcParams.update({'font.size': 22})
```

```python
generate_default_plot_env()
```

```python
why_df = pd.read_csv('../data/why_remove_mid.csv')
```

```python
fig, ax = plt.subplots(figsize=(20, 5))

unique_rel_days = list(pd.unique(why_df['move_relative_day']))

color = '#590059'
sns.pointplot(x='move_relative_day', y='mean',
              data=why_df, ci=None,
              #size=5, aspect=3, 
              scale=1.5,
              color=color, 
              ax=ax)

x_loc = [unique_rel_days.index(val) for val in why_df['move_relative_day']]
ax.errorbar(x=x_loc, y=why_df['mean'], 
            yerr=[why_df['bottom_ci'], why_df['top_ci']], 
            linestyle="None", ecolor=color,
            elinewidth=4.0);

if len(ax.xaxis.get_ticklabels()) % 2 == 1:
    left_line = len(ax.xaxis.get_ticklabels())/2 - 0.5
else:
    left_line = int(np.floor(len(ax.xaxis.get_ticklabels())/2 - 0.5))

right_line = left_line + 5
left_line -= 5

ax.axvline(x=left_line, color='k', linestyle='--', linewidth=1)
ax.axvline(x=right_line, color='k', linestyle='--', linewidth=1)
ax.axvspan(left_line, right_line, alpha=0.1, color='dimgray')

ax.text(0.49, 0.9, 'Filtered Out', transform=ax.transAxes, 
        ha='center', va='center', rotation=0,
        color='k', alpha=0.7, fontsize=30);

ax.set_ylim(bottom=5000, top=7000);
ax.grid(False)
ax.set_yticks(range(5000, 7001, 250))
ax.set_xticklabels([i if i % 2 == 0 else '' for i in range(-20, 21)])
# ax.set_title('Moving to Similarly Walkable Locations');
ax.set(xlabel=r'Days from Move $\left(t - t_{move}\right)$', ylabel='Daily Steps');

for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(20)

plt.tight_layout()

plt.savefig('../output/fig9.pdf', dpi=300)
```
