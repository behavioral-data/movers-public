```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

def generate_default_plot_env():
    sns.set(font_scale=1.4, style="whitegrid")
    matplotlib.rcParams['pdf.fonttype'] = 42
    matplotlib.rcParams['ps.fonttype'] = 42
    matplotlib.rc('font', family='sans-serif') 
    matplotlib.rc('font', serif='Helvetica Neue') 
    matplotlib.rc('text', usetex='false') 
    matplotlib.rcParams.update({'font.size': 22})
    
    plt.rcParams['pdf.fonttype'] = 42
    plt.rcParams['ps.fonttype'] = 42
    plt.rc('font', family='sans-serif') 
    plt.rc('font', serif='Helvetica Neue') 
    plt.rc('text', usetex='false') 
    plt.rcParams.update({'font.size': 22})

# pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 200)
```

```python
generate_default_plot_env()
```

```python
new_frac_pop_MVPAs_df = pd.read_csv("../data/new_frac_pop_MVPAs.csv", header=[0, 1])
new_frac_pop_MVPAs_df.columns = ["delta_ws"] + list(new_frac_pop_MVPAs_df.columns[1:])
```

```python
def plot_quantity(df, colname, ylabel, xlabel, color='#590059'):
    fig, ax = plt.subplots(figsize=(7.5, 5))
    ax.fill_between(x=df['delta_ws'], 
                    y1=df[colname, 'percentile_2.5'],
                    y2=df[colname, 'percentile_97.5'],
                    alpha=0.2,
                    color=color);
    ax.plot(df['delta_ws'], df[colname, 'mean'], color=color);

    ax.set_xlabel(xlabel);
    ax.set_ylabel(ylabel);

    # for label in ax.xaxis.get_ticklabels()[1::2]:
    #     label.set_visible(False)

    plt.tight_layout()
```

```python
def highlight_point(x, y, text, ax, color='k', 
                    arrow_end=None, xytext=(-20, -40), anno=True, 
                    connectionstyle='arc3'):
    
    arrow_end = (0, xytext[1]) if arrow_end is None else arrow_end
    
    if anno:
        ax.annotate('',
                    xy=(x, y), xycoords='data',
                    xytext=arrow_end, textcoords='offset pixels',
                    size=15,
                    arrowprops=dict(arrowstyle="->", 
                                    color=color,
                                    lw=1.,
                                    shrinkA=10, shrinkB=5,
                                    connectionstyle=connectionstyle),
                    color=color)
        
        ax.annotate(text,
                    xy=(x, y), xycoords='data',
                    xytext=xytext, textcoords='offset pixels',
                    size=15, color=color)
        
#         ax.annotate("%s --> %s" % (from_loc, to_loc),
#                     xy=(vals.iloc[0, 0], vals.iloc[0, 1]), xycoords='data',
#                     xytext=xytext, textcoords='offset pixels',
#                     size=15,
#                     arrowprops=dict(arrowstyle="->", 
#                                     color=color,
#                                     lw=1.,
#                                     shrinkA=5, shrinkB=5,
#                                     connectionstyle=connectionstyle),
#                     color=color)

plot_quantity(new_frac_pop_MVPAs_df, 'new_frac_pop_MVPAs', 
              'Estimated Fraction of Population with\n150+ Minutes of MVPA per Week', 
              'Simulated Increase in U.S. Walkability');
# manipulate
vals = plt.gca().get_yticks() 
plt.gca().set_yticklabels(['{:,.0%}'.format(x) for x in vals])

highlight_point(0, 0.185, 
                'status quo;\nno change\nin walkability', 
                plt.gca(),
                arrow_end=(50, 170), 
                xytext=(10, 170))
plt.xlim(0, 103);
plt.yticks(np.arange(0.15, 0.36, 0.05))
plt.grid(False)
plt.savefig('../output/fig3f.pdf', dpi=300)
```
