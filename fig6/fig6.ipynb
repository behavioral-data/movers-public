```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

def generate_default_plot_env():
    sns.set(font_scale=1.4, style="whitegrid")
    matplotlib.rcParams['pdf.fonttype'] = 42
    matplotlib.rcParams['ps.fonttype'] = 42
    matplotlib.rc('font', family='sans-serif') 
    matplotlib.rc('font', serif='Helvetica Neue') 
    matplotlib.rc('text', usetex='false') 
    matplotlib.rcParams.update({'font.size': 22})
    
    plt.rcParams['pdf.fonttype'] = 42
    plt.rcParams['ps.fonttype'] = 42
    plt.rc('font', family='sans-serif') 
    plt.rc('font', serif='Helvetica Neue') 
    plt.rc('text', usetex='false') 
    plt.rcParams.update({'font.size': 22})
```

```python
generate_default_plot_env()
```

```python
binned_moves_matrix = np.loadtxt("../data/binned_moves_matrix.txt")
```

```python
plt.figure()
g = sns.heatmap(binned_moves_matrix, annot=True, fmt='.0f', center=0.0, 
                cmap="RdBu_r", annot_kws={"size": 14})
g.set_ylabel('Walkability Quintile Before Move')
g.set_xlabel('Walkability Quintile After Move')
g.set_yticklabels([str(val) for val in range(1,5+1)])
g.set_xticklabels([str(val) for val in range(1,5+1)])
g.invert_yaxis()

figure = plt.gcf() # get current figure
figure.set_size_inches(8, 6)
plt.tight_layout()
plt.savefig('../output/fig6.pdf', dpi=300)
```
