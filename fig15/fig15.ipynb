```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

from tqdm import trange, tqdm
import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import json
sns.set_style("whitegrid")

import statsmodels.formula.api as smf
from scipy.stats.stats import pearsonr

def generate_default_plot_env():
    sns.set(font_scale=1.4, style="whitegrid")
    matplotlib.rcParams['pdf.fonttype'] = 42
    matplotlib.rcParams['ps.fonttype'] = 42
    matplotlib.rc('font', family='sans-serif') 
    matplotlib.rc('font', serif='Helvetica Neue') 
    matplotlib.rc('text', usetex='false') 
    matplotlib.rcParams.update({'font.size': 22})
    
    plt.rcParams['pdf.fonttype'] = 42
    plt.rcParams['ps.fonttype'] = 42
    plt.rc('font', family='sans-serif') 
    plt.rc('font', serif='Helvetica Neue') 
    plt.rc('text', usetex='false') 
    plt.rcParams.update({'font.size': 22})
    
global_min_steps = 500
global_max_steps = 50000
global_min_points = 10
```

```python
generate_default_plot_env()
```

```python
walkscore_moves_county_uneq_df = pd.read_csv("../data/steps_vs_binned_ws_income.csv")
```

```python
diff_col_name = 'diff_t1k_minc'

(_, bins) = pd.qcut(walkscore_moves_county_uneq_df[diff_col_name], 3, retbins=True)
bins[0] = -np.inf
bins[-1] = np.inf
bins = np.round(np.array(bins))
print(diff_col_name, bins)
walkscore_moves_county_uneq_df[diff_col_name + '_bin'] = pd.cut(walkscore_moves_county_uneq_df[diff_col_name], 
                                                                bins, right=False)

fig, ax = plt.subplots(figsize=(10, 5))
walkscore_moves_county_uneq_df.sort_values(['ws_diff', diff_col_name], inplace=True)
colors = ['#d77e2d', '#590059', '#3f6d32']
ax.axvline(x=2, color='#b0b0b0', linewidth=0.8, zorder=-1)
sns.pointplot(y='steps_diff', x='ws_diff_bin', data=walkscore_moves_county_uneq_df, 
              ax=ax, hue=diff_col_name + '_bin', dodge=0.2,
              palette=colors)

for idx in range(len(bins)-1):
    low_diff_val, high_diff_val = bins[idx], bins[idx+1]
    plot_df = walkscore_moves_county_uneq_df[(walkscore_moves_county_uneq_df[diff_col_name] >= low_diff_val) & 
                                             (walkscore_moves_county_uneq_df[diff_col_name] < high_diff_val)]

    sns.regplot(x=2.5 + (plot_df['ws_diff'] - 16)/32, y=plot_df['steps_diff'], 
                x_bins=range(-2, 2+1, 1), scatter=False, ci=None, ax=ax, color=colors[idx], 
                line_kws={'linestyle': '--', 'linewidth': 1.5, 'color':colors[idx], 'alpha': 0.5})

plt.setp(ax.get_xticklabels(), rotation=0)

ax.set_xlabel('$\Delta$ Walkability')
ax.set_ylabel('$\Delta$ Daily Steps')

leg_handles, labels = ax.get_legend_handles_labels()
legend = ax.legend(leg_handles, [r'(-$\infty$, -6250)', r'[-6250, 6954)', r'[6954, $\infty$)'], 
                   title='$\Delta$ Median County Household\nIncome After Relocation', loc='best')
legend.get_title().set_ha("center")
# legend.get_title().set_position((120, 0))

legend.get_title().set_fontsize('16.5') #legend 'Title' fontsize
plt.setp(plt.gca().get_legend().get_texts(), fontsize='15') #legend 'list' fontsize

for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(20)
    
ax.grid(False)

plt.tight_layout()
plt.savefig("../output/fig15.pdf", dpi=300)
```
